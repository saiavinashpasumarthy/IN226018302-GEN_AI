```python
# ============================================================
# IMPORTS
# ============================================================

!pip install -qU langchain-openai langchain
from dotenv import load_dotenv

from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_core.messages import HumanMessage
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain

from langchain.agents import initialize_agent, Tool
from langchain.tools import tool

from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
```

```python
# ============================================================
# LOAD ENVIRONMENT VARIABLES
# ============================================================

# Create a .env file and add:
# OPENAI_API_KEY=your_api_key_here

load_dotenv()
```

```python
# ============================================================
# INITIALIZE LLM
# ============================================================

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0.7
)
```

```python
# ============================================================
# 1. BASIC LLM CALL
# ============================================================

response = llm.invoke([
    HumanMessage(content="Explain LangChain in simple terms.")
])

print(response.content)
```

```python
# ============================================================
# 2. PROMPT TEMPLATE EXAMPLE
# ============================================================

template = PromptTemplate.from_template(
    "Explain {topic} in simple terms for a beginner."
)

prompt = template.invoke({
    "topic": "Vector Databases"
})

print(prompt.text)
```

```python
# ============================================================
# 3. SIMPLE CHAIN EXAMPLE
# ============================================================

prompt_template = ChatPromptTemplate.from_template(
    "Write a short explanation about {topic}"
)

model = ChatOpenAI(
    model="gpt-4o-mini"
)

parser = StrOutputParser()

chain = prompt_template | model | parser

result = chain.invoke({
    "topic": "LangChain Agents"
})

print(result)
```

```python
# ============================================================
# 4. MEMORY EXAMPLE
# ============================================================

memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)

conversation.predict(
    input="My name is Avinash"
)

response = conversation.predict(
    input="What is my name?"
)

print(response)
```

```python
# ============================================================
# 5. TOOL EXAMPLE
# ============================================================

@tool
def get_weather(city: str):
    """
    Returns weather information for a city.
    """
    return f"The weather in {city} is sunny."

weather_response = get_weather.invoke("Mumbai")

print(weather_response)
```

```python
# ============================================================
# 6. AGENT WITH TOOL
# ============================================================

@tool
def calculator(expression: str):
    """
    Performs mathematical calculations.
    """
    return eval(expression)

tools = [
    Tool(
        name="Calculator",
        func=calculator,
        description="Performs mathematical calculations"
    )
]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent="zero-shot-react-description",
    verbose=True
)

agent_response = agent.invoke(
    "What is 45 * 12?"
)

print(agent_response)
```

```python
# ============================================================
# 7. DOCUMENT LOADER EXAMPLE
# ============================================================

# Make sure research_paper.pdf exists in your project folder

try:
    loader = PyPDFLoader("research_paper.pdf")

    documents = loader.load()

    print(documents[0].page_content[:500])

except Exception as e:
    print("PDF not found or error loading document.")
    print(e)
```

```python
# ============================================================
# 8. VECTOR STORE / FAISS EXAMPLE
# ============================================================

texts = [
    "LangChain helps build LLM applications",
    "Vector databases enable semantic search",
    "Agents can use tools dynamically"
]

embeddings = OpenAIEmbeddings()

db = FAISS.from_texts(
    texts,
    embeddings
)

results = db.similarity_search(
    "What helps in semantic retrieval?"
)

print(results[0].page_content)
```

```python
# ============================================================
# 9. COMPLETE WORKFLOW EXAMPLE
# ============================================================

workflow_prompt = ChatPromptTemplate.from_template(
    """
    You are an AI tutor.

    Explain the concept of {topic}
    in simple and beginner-friendly terms.
    """
)

workflow_chain = workflow_prompt | llm | parser

workflow_result = workflow_chain.invoke({
    "topic": "Retrieval-Augmented Generation"
})

print(workflow_result)
```
