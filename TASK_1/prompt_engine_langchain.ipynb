# GenAI Assignment – Prompt Engineering using LangChain

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
```

## Task 1: Replace Hardcoded Prompts

In this task, we convert a static prompt into a reusable LangChain PromptTemplate.

```python
# PromptTemplate instead of f-string
template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Test
print(template_1.format(topic="Artificial Intelligence"))
```

## Task 2: Multi-Input Prompt System

This task demonstrates how to handle multiple dynamic inputs such as topic, audience, and tone.

```python
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test Cases
print(multi_input_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_input_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

## Task 3: Prompt Variations Engine

Here, we create multiple prompt templates for different use cases like teaching, interview, and storytelling.

```python
# Teaching Template
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

# Interview Template
interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

# Storytelling Template
story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

# Run all
topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System

This task uses role-based prompting with system and user messages.

```python
def get_chat_prompt(role, topic):
    if role == "teacher":
        system_msg = "You are a helpful teacher."
    elif role == "interviewer":
        system_msg = "You are an interviewer asking questions."
    elif role == "motivator":
        system_msg = "You are a motivational speaker."
    else:
        system_msg = "You are a helpful assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_msg),
        ("user", "Talk about {topic}")
    ])

    return chat_template.format_messages(topic=topic)

# Test
messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
```

## Task 5: Input Validation Layer

We validate user inputs to ensure correct values for audience and tone.

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Setting default = beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Setting default = formal")
        tone = "formal"

    return audience, tone

# Test
print(validate_inputs("kids", "fun"))
```

## Task 6: Prompt Generator App

A function that dynamically generates prompts based on user inputs and selected style.

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = "Explain {topic} for {audience} in a {tone} teaching style"
    elif style == "interview":
        template = "Ask interview questions about {topic} for {audience} in a {tone} tone"
    elif style == "storytelling":
        template = "Explain {topic} for {audience} in a {tone} storytelling style"
    else:
        template = "Explain {topic} for {audience} in a {tone} tone"

    prompt_template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=template
    )

    return prompt_template.format(topic=topic, audience=audience, tone=tone)

# Test
print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability Test

This demonstrates how one template can be reused with multiple inputs.

```python
reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} to {audience}"
)

# Different inputs, same template
inputs = [
    ("AI", "beginners"),
    ("Python", "kids"),
    ("Cybersecurity", "students"),
    ("Blockchain", "developers"),
    ("Data Science", "analysts")
]

for topic, audience in inputs:
    print(reusable_template.format(topic=topic, audience=audience))
```

## Final Pipeline Flow

User Input → Validation → Prompt Template → Output

This shows how all components are integrated together.

```python
def full_pipeline(topic, audience, tone, style):
    # Step 1: Validate
    audience, tone = validate_inputs(audience, tone)

    # Step 2: Generate Prompt
    result = generate_prompt(topic, audience, tone, style)

    return result

# Run Pipeline
print(full_pipeline("AI", "beginner", "fun", "teaching"))
```
